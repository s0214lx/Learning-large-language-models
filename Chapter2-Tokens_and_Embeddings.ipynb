```python
from transformers import AutoModelForCausalLM,AutoTokenizer

# å®šä¹‰é¢œè‰²åˆ—è¡¨
colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

# å®šä¹‰show_tokenså‡½æ•°
def show_tokens(sentence, tokenizer_name):
    # ä»æŒ‡å®šçš„æ¨¡å‹åç§°åŠ è½½å¯¹åº”çš„åˆ†è¯å™¨
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    # å°†å¥å­ç¼–ç æˆtoken idåˆ—è¡¨
    token_ids = tokenizer(sentence).input_ids
    # å¾ªç¯æ‰“å°token
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '    
        )

# æµ‹è¯•æ–‡æœ¬
text = """
English and CAPITALIZATION
ğŸµ é¸Ÿ
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
"""
```

```python
show_tokens(text, "bert-base-uncased")
```

```python
# åŸºäº embeddings çš„æ­Œæ›²æ¨ç†
import pandas as pd
# ç”¨äºä»ç½‘ç»œä¸Šä¸‹è½½æ–‡ä»¶æˆ–ç½‘é¡µå†…å®¹
from urllib import request

# æ‰“å¼€ä¸€ä¸ªurlé“¾æ¥ï¼Œè¯»å–è¿œç¨‹æœåŠ¡å™¨ä¸Šåä¸ºtrain.txtçš„æ•°æ®æ–‡ä»¶
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# è¯»å–æ–‡ä»¶ï¼Œå°†å­—èŠ‚æ•°æ®è§£ç æˆå­—ç¬¦ä¸²ï¼Œç„¶åæŒ‰è¡Œå°†æ•´ä¸ªæ–‡æœ¬åˆ†å‰²æˆä¸€ä¸ªå­—ç¬¦ä¸²åˆ—è¡¨ã€‚å¹¶è·³è¿‡å‰ä¸¤è¡Œ
lines = data.read().decode('utf-8').split('\n')[2:]

# å¯¹æ¯ä¸€è¡Œè¿›è¡Œå¤„ç†ï¼Œå»é™¤ç»“å°¾çš„ç©ºç™½å­—ç¬¦ï¼ŒæŒ‰ç©ºæ ¼åˆ†å‰²æˆä¸€ä¸ªå­—ç¬¦ä¸²åˆ—è¡¨ï¼šæ¯ä¸ªå…ƒç´ æ˜¯ä¸€ä¸ªæ’­æ”¾åˆ—è¡¨ï¼ˆå«å¤šé¦–æ­Œï¼‰ï¼Œä»¥æ­Œæ›² ID å½¢å¼å­˜å‚¨çš„å­—ç¬¦ä¸²æ•°ã€‚å¹¶è¿‡æ»¤æ‰åªæœ‰ 1 é¦–æ­Œçš„æ’­æ”¾åˆ—è¡¨ï¼ˆä¸åˆ©äºè®­ç»ƒ Word2Vecï¼‰ã€‚
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]

# åŠ è½½æ­Œæ›²å…ƒæ•°æ®
songs_file = request.urlopen("https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt")
songs_file = songs_file.read().decode("utf-8").split("\n")
songs = [s.rstrip().split('\t') for s in songs_file]

# å°†songsåˆ—è¡¨è½¬æ¢æˆpandas.DataFrameï¼Œæ¯ä¸€è¡Œæ˜¯ä¸€é¦–æ­Œ
songs_df = pd.DataFrame(data=songs, columns = ['id','title','artist'])
# è®¾ç½®ç´¢å¼•åˆ—
songs_df = songs_df.set_index('id')
```

```python
# å¯¼å…¥woed2evcæ¨¡å‹ç±»ï¼Œç”¨äºè®­ç»ƒè¯å‘é‡
from gensim.models import Word2Vec

# ä½¿ç”¨å“Ÿå…³äºæ’­æ”¾åˆ—è¡¨ï¼ˆplaylistï¼‰æ•°æ®æ¥è®­ç»ƒWord2Vecæ¨¡å‹
model = Word2Vec(
    playlists, # è¾“å…¥æ•°æ®
    vector_size = 32, # æ¯ä¸ªè¯å‘é‡ä¼šè¢«ç¼–ç ä¸ºä¸€ä¸ª32ç»´çš„å‘é‡
    window = 20, # è®­ç»ƒæ—¶ï¼Œæ¨¡å‹ä¼šåœ¨ä¸€ä¸ªçª—å£å¤§å°ä¸º20çš„èŒƒå›´å†…å¯»æ‰¾ä¸Šä¸‹æ–‡è¯
    negative = 50 , # è´Ÿé‡‡æ ·æ•°é‡ã€‚æ¯ä¸ªæ­£æ ·æœ¬ä¼šé…åˆ50ä¸ªè´Ÿæ ·æœ¬è¿›è¡Œè®­ç»ƒã€‚è´Ÿæ ·æœ¬ç”¨äºæé«˜è®­ç»ƒæ•ˆç‡
    min_count = 1, # åªè¦æŸä¸ªtokenå‡ºç°è‡³å°‘1æ¬¡å°±ä¼šè¢«è€ƒè™‘è¿›è®­ç»ƒæ•°æ®ä¸­
    workers = 4, # ä½¿ç”¨4ä¸ªçº¿ç¨‹è¿›è¡Œå¹¶è¡Œè®­ç»ƒï¼ŒåŠ å¿«é€Ÿåº¦
)
```

```python
# æƒ³è¦æŸ¥è¯¢çš„ç›®æ ‡æ­Œæ›²ID
song_id = 2172
# mode.wv:è®¿é—®word2evcæ¨¡å‹ä¸­è®­ç»ƒå¥½çš„è¯å‘é‡
# most_similar:åœ¨è¯å‘é‡ç©ºé—´ä¸­ï¼Œæ‰¾åˆ°ä¸ç»™å®šå‘é‡æœ€ç›¸è¿‘çš„å‘é‡ï¼ˆé€šè¿‡ä½™å¼¦ç›¸ä¼¼åº¦cosine similarityè®¡ç®—ï¼‰
# æ³¨æ„ï¼šword2evcçš„most_similaréœ€è¦è¾“å…¥è¯ï¼ˆtokenï¼‰ï¼Œè€Œplayistsä¸­çš„æ¯é¦–æ­Œæ˜¯ä¸€ä¸ªå­—ç¬¦ä¸²ï¼Œå› æ­¤éœ€è¦å°†song_idè½¬ä¸ºå­—ç¬¦ä¸²ç±»å‹çš„
model.wv.most_similar(positive=str(song_id))
```

```python
import numpy as np

# å®šä¹‰æ¨èå‡½æ•°
def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive = str(song_id),topn=5)
    )[:,0]
    return songs_df.iloc[similar_songs]

print_recommendations(2171)
```
